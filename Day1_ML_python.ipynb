# Implement Simple linear regression
a.) Choose a feature(ex-square footage) as an independent variable(X) and house prices as the dependant variable(Y)
b.) Implement simple linear regression model using sklearn to predict house prices based on the selected feature
c.)Visualize the data and the regression line

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
#train the data
print("Training Started.......\n")
print()
lin_reg.fit(ctrain,etrain)
print('Training completed.....')
print()
#test the data
print('Testing invoked......\n')
epredict=lin_reg.predict(ctest)
print('Predicted price\n',epredict)
print('\nTesting is also completed....\n')
```

```python
etrain
```

```python
etest
```

```python
diff=etest-epredict
diff
```

```python
#Visualise the training dataset
plt.scatter(ctrain,etrain,color='red')
plt.plot(ctrain,lin_reg.predict(ctrain),color='green')
plt.title('Prices according to squarefoot of the house(training dataset)')
plt.xlabel('sqft')
plt.ylabel('price')
plt.show()
```

```python
#Visualise the test dataset
plt.scatter(ctest,etest,color='green')
plt.plot(ctest,lin_reg.predict(ctest),color='red')
plt.title('Prices according to squarefoot of the house(testing dataset)')
plt.xlabel('sqft')
plt.ylabel('price')
plt.show()
```

# Evaluate the Simple linear regression model
a.)Use scikit learn to calculate the R-squared value to assess the goodness of fit
b.)Interpret the R-squared value and discuss the model's performance

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import explained_variance_score

print('Mean Squared Error ; \t',mean_squared_error(etest,epredict))
print()
print('RMSE:\t',np.sqrt(mean_squared_error(etest,epredict)))
print()
print('Variance Score:\t',explained_variance_score(etest,epredict))

#Coefficient of determination- R_square
print()
print('R-Square:\t',r2_score(etest,epredict))
```
