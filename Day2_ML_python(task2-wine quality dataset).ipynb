# Build Base Regression Model and Observe the performance

```python
print('**********Base Regression***********')
from sklearn.linear_model import LinearRegression
base_reg=LinearRegression()
#train the data
print("Training Started.......\n")
print()
base_reg.fit(xtrain,ytrain)
print('Training completed.....')
print()
#test the data
print('Testing invoked......\n')
ypredict=base_reg.predict(xtest)
print('Predicted price\n',ypredict)
print('\nTesting is also completed....\n')

```

```python
print('********Performance metrics of Linear Regression model*****************')
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import explained_variance_score

print("Coefficient-M:\t", base_reg.coef_)
print()
print('Intercept-C:\t', base_reg.intercept_)
print()

print('Mean Squared Error ; \t',mean_squared_error(ytest,ypredict))
print()
print('RMSE:\t',np.sqrt(mean_squared_error(ytest,ypredict)))
print()
print('Variance Score:\t',explained_variance_score(ytest,ypredict))
print()
print("Base Train_score:\t", base_reg.score(xtrain,ytrain))
print()
print("Base Test_score:\t", base_reg.score(xtest,ytest))
print()
print("Base R-Square:\t",r2_score(ytest,ypredict))
```

# Build polynomial based linear regression model and observe the performance with optimum degree/order of polynomial features

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_features=PolynomialFeatures(degree=2)
x_poly=poly_features.fit_transform(x)
y_poly=poly_features.fit_transform(y)

#Call linear regression and assign to poly_reg variable

poly_reg=LinearRegression()
poly_reg.fit(x_poly,y)
print("Coefficient-M:\t", poly_reg.coef_)
print()
print('Intercept-C:\t', poly_reg.intercept_)
print()
```

```python
#Visualize base polynomial based regression line
plt.scatter(x,y)
plt.plot(x,poly_reg.predict(x_poly),color='green')
plt.title('Wine quality-polynomial based regression')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.show()
```
